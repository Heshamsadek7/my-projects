```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
# تحميل البيانات
df = pd.read_csv('/home/hesham/NMG/8-25/new/work permations.csv', encoding='utf-8')
```

```python
df.columns
```

```python
df=df[['الشهر', 'التاريخ', 'اليوم', 'إسم الشركة', 'المشرف التنفيزى ',
       'نوع العمل', 'مواقف التصريح', 'ساعة الدخول ', 'ساعة الخروج',
       #'إجمالي عدد الساعات',
       'الموقع', 'إجمالي ساعات العمل',
        # 'المصدر', 'عمود1',
       'مشرف السلامة', 
       #'تقييم العملية', 'عمود2', 'عمود3', 'عمود4', 'عمود5',
       #'عمود6', 'عمود7', 'عمود8', 'عمود9', 'عمود10',
       # 'Column1'
       ]]
```

```python
df.columns
```

```python
# إزالة المسافات والفواصل إذا كانت موجودة
df['إجمالي ساعات العمل'] = df['إجمالي ساعات العمل'].str.replace('٫', '.', regex=False)


# ثم التحويل إلى numeric ثم float
df['إجمالي ساعات العمل'] = df['إجمالي ساعات العمل'].astype(float)
```

```python
# فصل البيانات حسب الشهر
august_permations = df[df['الشهر'] == 'August-25']
july_permations = df[df['الشهر'] == 'يوليو-25']
```

```python
df['الشهر']=pd.to_datetime(df['الشهر'], errors='coerce')
df['التاريخ']=pd.to_datetime(df['التاريخ'], errors='coerce')
df['اليوم']=pd.to_datetime(df['اليوم'], errors='coerce')
df['ساعة الدخول ']=pd.to_datetime(df['ساعة الدخول '], errors='coerce')
df['ساعة الخروج']=pd.to_datetime(df['ساعة الخروج'], errors='coerce')
```

```python
df['الموقع'].value_counts().reset_index()
```

```python
df.loc[df['الموقع'] != 'دويتش '].groupby('الموقع')['إجمالي ساعات العمل'].sum().sort_values(ascending=False).reset_index()
```

```python
By_location=df.loc[df['الموقع'] != 'دويتش '].groupby('الموقع')['إجمالي ساعات العمل'].sum().sort_values(ascending=False)
```

```python
# قم بإنشاء DataFrame بدلاً من Series
By_location = df.loc[df['الموقع'] != 'دويتش '].groupby('الموقع')['إجمالي ساعات العمل'].sum().sort_values(ascending=False).reset_index()

# الآن استخدم DataFrame كاملاً في الرسم
px.bar(data_frame=By_location, 
       x='الموقع', 
       y='إجمالي ساعات العمل',
       color='الموقع',text_auto=True,title='تأمين أعمال المقاولات الخارجية حسب الموقع بالساعات')
```

```python
result = df.loc[df['الموقع'] != 'دويتش '].groupby(['مشرف السلامة', 'الموقع'])['إجمالي ساعات العمل'].sum().unstack().fillna(0)
result.loc[result.sum(axis=1).sort_values(ascending=False).index]
```

```python
hse=result.loc[result.sum(axis=1).sort_values(ascending=False).index]
```

```python
px.bar(data_frame=hse,text_auto=True,title='تأمين أعمال المقاولين الخارجيين بالساعات')
```
