```python
df.loc[df['Location'] != 'Deutsch'].groupby('Location')['Total Hours'].sum().sort_values(ascending=False)
```

```python
result = df.loc[df['Location'] != 'Deutsch'].groupby(['Safety Supervisor', 'Location'])['Total Hours'].sum().unstack().fillna(0)
result.loc[result.sum(axis=1).sort_values(ascending=False).index]
```

```python
plt.figure(figsize=(14, 10))

ax = result.loc[result.sum(axis=1).sort_values(ascending=False).index].plot(
    kind='barh', 
    color=MY_colors,
    width=1.5,
    edgecolor='black',
    linewidth=0.5
)

# إخفاء التسميات للقيم الصفرية
for container in ax.containers:
    labels = [f'{int(val)}' if val > 0 else '' for val in container.datavalues]
    ax.bar_label(container, labels=labels, label_type='center', padding=3, fontsize=10)

# تعديل حدود المحور لإخفاء المساحة الفارغة
plt.xlim(left=0, right=result.values.max() * 1.1)  # زيادة بنسبة 10% لأعلى قيمة

plt.title('Safety Supervisor', color="#B4056B", fontsize=16, fontweight='bold')
plt.xlabel('Total Hours', fontsize=12)
plt.ylabel('Safety Supervisor', fontsize=12)
plt.legend(title='Location', title_fontsize=12, fontsize=10)

ax.set_facecolor("#f5e1cf")
plt.grid(axis='x', alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()
```

```python
location_TotalHours=df.loc[df['Location'] != 'Deutsch'].groupby('Location')['Total Hours'].sum().sort_values(ascending=False)
location_TotalHours
```
