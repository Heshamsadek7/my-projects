```python
# البحث عن انسكاب/تسريب سيليكون
انسكاب_سليكون = (
    ncr[
        ncr['وصف حالة عدم التطابق\nDescription of non conformance']
        .str.contains('انسكاب.*سليكون|تسريب.*سليكون|سليكون.*انسكاب|سليكون.*تسريب', na=False)
    ]
)

انسكاب_سليكون[[
    'وصف حالة عدم التطابق\nDescription of non conformance', 
    'الإدارة المتواجد بها الحالة', 
    'المتابعة\nFollow Up'
]]
```

```python
app = Dash(__name__)

# تحضير البيانات - استخراج نوع الانسكاب من الوصف
def استخراج_نوع_الانسكاب(وصف):
    وصف = str(وصف).lower()
    if 'مياه' in وصف or 'ماء' in وصف:
        return 'مياه'
    elif 'زيت' in وصف:
        return 'زيت'
    elif 'حبر' in وصف:
        return 'حبر'
    elif 'سليكون' in وصف:
        return 'سليكون'
    else:
        return None  # إرجاع None للحالات غير المرغوبة

# إنشاء نسخة من البيانات مع عمود نوع الانسكاب
ncr_مع_الانسكاب = ncr.copy()
ncr_مع_الانسكاب['نوع_الانسكاب'] = ncr_مع_الانسكاب['وصف حالة عدم التطابق\nDescription of non conformance'].apply(استخراج_نوع_الانسكاب)

# تصفية البيانات لإزالة الحالات غير المرغوبة (None)
ncr_انسكابات_فقط = ncr_مع_الانسكاب.dropna(subset=['نوع_الانسكاب'])

# الحصول على الأشهر المتاحة من البيانات المصفاة
available_months = sorted([int(month) for month in ncr_انسكابات_فقط['شهر'].dt.month.unique()])

# إنشاء خيارات القائمة المنسدلة
dropdown_options = [{'label': 'الكل (المجموع الكلي)', 'value': 'all'}] + [{'label': f'شهر {month}', 'value': month} for month in available_months]

app.layout = html.Div([
    html.H1("داشبورد تحليل أنواع الانسكابات", 
             style={"color": "#2E86AB", "textAlign": "center", "marginBottom": "20px"}),
    
    html.Div([
        html.Label("اختر الشهر:", style={"fontWeight": "bold", "marginRight": "10px"}),
        dcc.Dropdown(
            id="month-dropdown", 
            options=dropdown_options,
            value='all',
            style={"width": "250px"}
        ),
    ], style={"display": "flex", "justifyContent": "center", "alignItems": "center", "marginBottom": "30px"}),
    
    html.Div([
        dcc.Graph(id="leak-type-chart"),
        dcc.Graph(id="leak-department-chart")
    ], style={"display": "grid", "gridTemplateColumns": "1fr 1fr", "gap": "20px"}),
    
    html.Div(id="stats-cards", style={"marginTop": "30px"})
])

@app.callback(
    [Output("leak-type-chart", "figure"),
     Output("leak-department-chart", "figure"),
     Output("stats-cards", "children")],
    [Input("month-dropdown", "value")]
)
def update_dashboard(selected_month):
    # تصفية البيانات بناء على الشهر
    if selected_month == 'all':
        filtered_df = ncr_انسكابات_فقط
        title_suffix = ' - المجموع الكلي'
    else:
        filtered_df = ncr_انسكابات_فقط[ncr_انسكابات_فقط['شهر'].dt.month == selected_month]
        title_suffix = f' - شهر {selected_month}'
    
    # الرسم البياني 1: توزيع أنواع الانسكابات
    fig1 = px.histogram(
        filtered_df,
        x='نوع_الانسكاب',
        color='نوع_الانسكاب',
        text_auto=True,
        title=f'توزيع أنواع الانسكابات{title_suffix}',
        color_discrete_map={
            'مياه': '#3498db',
            'زيت': '#e74c3c', 
            'حبر': '#2c3e50',
            'سليكون': '#f39c12'
        }
    )
    
    fig1.update_layout(
        xaxis_title="نوع الانسكاب",
        yaxis_title="عدد الحالات",
        showlegend=False,
        xaxis=dict(tickangle=0)
    )
    
    # الرسم البياني 2: توزيع الانسكابات حسب الإدارة
    fig2 = px.histogram(
        filtered_df,
        x='الإدارة المتواجد بها الحالة',
        color='نوع_الانسكاب',
        text_auto=True,
        barmode='stack',
        title=f'الانسكابات حسب الإدارة{title_suffix}',
        color_discrete_map={
            'مياه': '#3498db',
            'زيت': '#e74c3c',
            'حبر': '#2c3e50',
            'سليكون': '#f39c12'
        }
    )
    
    fig2.update_layout(
        xaxis_title="الإدارة",
        yaxis_title="عدد الحالات",
        legend_title="نوع الانسكاب",
        xaxis=dict(tickangle=45)
    )
    
    # إنشاء بطاقات الإحصائيات
    total_cases = len(filtered_df)
    leak_types = filtered_df['نوع_الانسكاب'].value_counts()
    
    stats_cards = html.Div([
        html.Div([
            html.H4("إجمالي حالات الانسكاب", style={"color": "#2E86AB", "margin": "0"}),
            html.H2(f"{total_cases}", style={"color": "#2E86AB", "margin": "0"})
        ], style={"flex": "1", "textAlign": "center", "padding": "15px", "background": "#f8f9fa", "margin": "5px", "borderRadius": "8px"}),
        
        html.Div([
            html.H4("حالات تسريب المياه", style={"color": "#3498db", "margin": "0"}),
            html.H2(f"{leak_types.get('مياه', 0)}", style={"color": "#3498db", "margin": "0"})
        ], style={"flex": "1", "textAlign": "center", "padding": "15px", "background": "#f8f9fa", "margin": "5px", "borderRadius": "8px"}),
        
        html.Div([
            html.H4("حالات تسريب الزيت", style={"color": "#e74c3c", "margin": "0"}),
            html.H2(f"{leak_types.get('زيت', 0)}", style={"color": "#e74c3c", "margin": "0"})
        ], style={"flex": "1", "textAlign": "center", "padding": "15px", "background": "#f8f9fa", "margin": "5px", "borderRadius": "8px"}),
        
        html.Div([
            html.H4("حالات تسريب الحبر", style={"color": "#2c3e50", "margin": "0"}),
            html.H2(f"{leak_types.get('حبر', 0)}", style={"color": "#2c3e50", "margin": "0"})
        ], style={"flex": "1", "textAlign": "center", "padding": "15px", "background": "#f8f9fa", "margin": "5px", "borderRadius": "8px"}),
        
        html.Div([
            html.H4("حالات تسريب السليكون", style={"color": "#f39c12", "margin": "0"}),
            html.H2(f"{leak_types.get('سليكون', 0)}", style={"color": "#f39c12", "margin": "0"})
        ], style={"flex": "1", "textAlign": "center", "padding": "15px", "background": "#f8f9fa", "margin": "5px", "borderRadius": "8px"})
    ], style={"display": "flex", "justifyContent": "center", "maxWidth": "1200px", "margin": "0 auto",})
    
    return fig1, fig2, stats_cards

if __name__ == "__main__":
    print("🚀 تشغيل داشبورد الانسكابات...")
    print(f"📊 إجمالي حالات الانسكاب الحقيقية: {len(ncr_انسكابات_فقط)}")
    print("📈 توزيع أنواع الانسكابات:")
    print(ncr_انسكابات_فقط['نوع_الانسكاب'].value_counts())
    app.run(mode='inline', port=8076, debug=True)
    # http://localhost:8076/
```
