```python
from dash import html, dcc, Dash
from jupyter_dash import JupyterDash
from dash.dependencies import Input, Output
from dash.exceptions import PreventUpdate
import plotly.express as px
import pandas as pd
import seaborn as sns

# استخدم JupyterDash إذا كنت تعمل في Jupyter
app =Dash(__name__)

# إنشاء خيارات الدروب داون بشكل صحيح
months = sorted(ncr['شهر'].dt.month.unique())
dropdown_options = [{'label': f'شهر {month}', 'value': month} for month in months]

app.layout = html.Div([
    html.H3("حالات المتابعة حسب الإدارة", style={"color": "red"}),
    dcc.Dropdown(
        id="dropdown", 
        options=dropdown_options,
        value=months[0] if months else None,  # قيمة افتراضية
        placeholder="اختر الشهر"
    ),
    dcc.Graph(id="visual")
])

@app.callback(Output("visual", "figure"), Input("dropdown", "value"))
def update_graph(month):
    if month is None:
        raise PreventUpdate
    
    # تصفية البيانات للشهر المحدد
    filtered_df = ncr[ncr['شهر'].dt.month == month]
    
    figure = px.histogram(
        filtered_df,
        x='الإدارة المتواجد بها الحالة',
        color='المتابعة\nFollow Up',
        text_auto=True,
        barmode='group',
        title=f'حالات المتابعة (Opened/Closed) حسب الإدارة - شهر {month}'
    )
    return figure

if __name__ == "__main__":
    # إذا كنت في Jupyter
    app.run(mode="inline", port=8068)
    # http://localhost:8068
    
    # أو إذا كنت في سكريبت عادي
    # app.run(host='127.0.0.1', port=8068, debug=True)
```

```python
from dash import html, dcc, Dash
from jupyter_dash import JupyterDash
from dash.dependencies import Input, Output
from dash.exceptions import PreventUpdate
import plotly.express as px
import pandas as pd
import seaborn as sns

# استخدم JupyterDash إذا كنت تعمل في Jupyter
app = Dash(__name__)

# إنشاء خيارات الدروب داون بشكل صحيح مع إضافة "الكل"
months = sorted(ncr['شهر'].dt.month.unique())
dropdown_options = [{'label': 'الكل', 'value': 'all'}] + [{'label': f'شهر {month}', 'value': month} for month in months]

app.layout = html.Div([
    html.H3("حالات المتابعة حسب الإدارة", style={"color": "red"}),
    dcc.Dropdown(
        id="dropdown", 
        options=dropdown_options,
        value='all',  # قيمة افتراضية (الكل)
        placeholder="اختر الشهر"
    ),
    dcc.Graph(id="visual")
])

@app.callback(Output("visual", "figure"), Input("dropdown", "value"))
def update_graph(selected_value):
    if selected_value is None:
        raise PreventUpdate
    
    # تحديد البيانات بناء على الاختيار
    if selected_value == 'all':
        filtered_df = ncr  # جميع البيانات
        title = 'حالات المتابعة (Opened/Closed) حسب الإدارة - جميع الأشهر'
    else:
        filtered_df = ncr[ncr['شهر'].dt.month == selected_value]
        title = f'حالات المتابعة (Opened/Closed) حسب الإدارة - شهر {selected_value}'
    
    figure = px.histogram(
        filtered_df,
        x='الإدارة المتواجد بها الحالة',
        color='المتابعة\nFollow Up',
        text_auto=True,
        barmode='overlay',
        title=title
    )
    
    # تحسين إضافي للتنسيق
    figure.update_layout(
        xaxis_title="الإدارة",
        yaxis_title="عدد الحالات",
        legend_title="حالة المتابعة"
    )
    
    return figure

if __name__ == "__main__":
    # إذا كنت في Jupyter
    app.run(mode="inline", port=8069)
    
    # أو إذا كنت في سكريبت عادي
    # app.run(host='127.0.0.1', port=8068, debug=True)
    # http://localhost:8069/
    
```

```python
from dash import html, dcc, Dash
from jupyter_dash import JupyterDash
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd
import numpy as np

# app = JupyterDash(__name__)
app = Dash(__name__)

# معالجة np.int32 بشكل صحيح
available_months = sorted(ncr['شهر'].dt.month.unique())
dropdown_options = [{'label': 'الكل', 'value': 'all'}] 

for month in available_months:
    # تحويل np.int32 إلى int عادي للعرض، ولكن الاحتفاظ بالقيمة الأصلية للمقارنة
    month_int = int(month)
    dropdown_options.append({'label': f'شهر {month_int}', 'value': month})

app.layout = html.Div([
    html.H1("تقرير المتابعة بحسب فنيين إدارة السلامة", style={"color": "#2E86AB", "textAlign": "center"}),
    
    html.Div([
        dcc.Dropdown(
            id="month-dropdown", 
            options=dropdown_options,
            value='all',
            style={"width": "250px", "margin": "20px auto"}
        ),
    ]),
    
    dcc.Graph(id="follow-up-chart")
])

@app.callback(Output("follow-up-chart", "figure"), Input("month-dropdown", "value"))
def update_chart(selected_month):
    print(f"📅 الشهر المختار: {selected_month} (نوع: {type(selected_month)})")
    
    if selected_month == 'all':
        filtered_df = ncr
        title = 'المجموع الكلي لجميع الأشهر'
    else:
        # استخدام القيمة الأصلية (np.int32) للمقارنة
        filtered_df = ncr[ncr['شهر'].dt.month == selected_month]
        title = f'شهر {int(selected_month)}'
    
    print(f"📊 عدد الصفوف المصفاة: {len(filtered_df)}")
    
    # التأكد من وجود بيانات
    if len(filtered_df) == 0:
        return px.bar(title="لا توجد بيانات")
    
    fig = px.histogram(
        filtered_df,
        x='إسم الشخص القائم بالتبليغ',
        color='المتابعة\nFollow Up',
        barmode='overlay',
        text_auto=True,
        title=title
    )
    
    fig.update_layout(
        xaxis_title="الفنيين",
        yaxis_title="عدد الحالات", 
        xaxis_tickangle=45
    )
    
    return fig

if __name__ == "__main__":
    app.run(mode="inline", port=8071, debug=True)
# http://localhost:8071/
```
