```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
import numpy as np
```

```python
# تحميل البيانات
df = pd.read_csv('cleaned_ncr_data8-25.csv')
```

```python
# تنظيف البيانات
df['Month'] = df['Month'].str.strip()
```

```python
# Set the style for the plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# 1. Follow-up status by month
plt.figure(figsize=(10, 6))
status_by_month = pd.crosstab(df['Month'], df['Follow_Up'])
ax = status_by_month.plot(kind='bar', color=['#ff9999', '#66b3ff'])
plt.title('Follow-up Status by Month')
plt.xlabel('Month')
plt.ylabel('Number of Reports')
plt.legend(['Closed', 'Open'])
plt.xticks(rotation=45)
plt.tight_layout()
for container in ax.containers:
        ax.bar_label(container, fmt='%d', label_type='edge', padding=3, fontsize=9)
plt.show()
```

```python
# 2. Number of reports by department
plt.figure(figsize=(12, 6))
dept_counts = df['Department'].value_counts()
ax = sns.barplot(x=dept_counts.index, y=dept_counts.values, palette='viridis')
plt.title('Number of Reports by Department')
plt.xlabel('Department')
plt.ylabel('Number of Reports')
plt.xticks(rotation=45)
plt.tight_layout()
for container in ax.containers:
        ax.bar_label(container, fmt='%d', label_type='edge', padding=3, fontsize=9)
plt.show()
```

```python
# 3. Number of reports by reporter (top 10)
plt.figure(figsize=(12, 6))
reporter_counts = df['Reporter_Name'].value_counts().head(10)
ax = sns.barplot(x=reporter_counts.index, y=reporter_counts.values, palette='magma')
plt.title('Number of Reports by Reporter (Top 10)')
plt.xlabel('Reporter')
plt.ylabel('Number of Reports')
plt.xticks(rotation=45)
plt.tight_layout()
for container in ax.containers:
        ax.bar_label(container, fmt='%d', label_type='edge', padding=3, fontsize=9)
plt.show()
```

```python
# 4. Risk level by month
plt.figure(figsize=(10, 6))
risk_by_month = pd.crosstab(df['Month'], df['Risk_Level'])
ax = risk_by_month.plot(kind='bar', color=['#ff9999', '#ffcc99', '#ff6666'])
plt.title('Risk Level by Month')
plt.xlabel('Month')
plt.ylabel('Number of Reports')
plt.legend(['Medium', 'High'])
plt.xticks(rotation=45)
plt.tight_layout()
for container in ax.containers:
        ax.bar_label(container, fmt='%d', label_type='center', padding=3, fontsize=9)
plt.show()
```

```python
# 5. Source of reports
plt.figure(figsize=(8, 8))
source_counts = df['Source'].value_counts()
plt.pie(source_counts.values, labels=source_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Report Sources')
plt.axis('equal')
plt.show()
```

```python
# 7. Equipment analysis
plt.figure(figsize=(12, 6))
equipment_counts = df['Equipment_Tool'].value_counts().head(10)
ax = sns.barplot(x=equipment_counts.index, y=equipment_counts.values, palette='rocket')
plt.title('Most Reported Equipment/Tools (Top 10)')
plt.xlabel('Equipment/Tool')
plt.ylabel('Number of Reports')
plt.xticks(rotation=45)
for container in ax.containers:
        ax.bar_label(container, fmt='%d', label_type='edge', padding=3, fontsize=9)
plt.tight_layout()

plt.show()
```
