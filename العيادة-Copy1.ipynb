```python
df['الادارة التابع لها ']=df['الادارة التابع لها '].astype(str)
```

```python
dep=['التجليد', 'التنفيذ', 'الهندسية', 'الشئون الادارية ', 'حركة ركاب',
       'الشيت', '0', 'الاكاديمية', 'الويب', 'التجهيزات', 'الجودة ',
       'عمليات الانتاج', 'الامن', 'دشت', 'الاكاديمة',
       'السلامة والصحة المهنية ', 'حركة النقل', 'تعبئة وتغليف',
       'مراقبة الكاميرات ', 'الجودة', 'عمالة خارجية ', 'سلاسل الامداد',
       'التسليمات', 'تخطيط الانتاج']
```

```python
df['الادارة التابع لها '].unique()
```

```python
df['الادارة التابع لها ']=df['الادارة التابع لها ']\
    .replace({'الاكاديمة':'الاكاديمية','حركة النقل':'الشئون الادارية '
              ,'الجودة':'جودة','0':'عمالة خارجية ','الجودة ':'جودة'})
```

```python
ax=df.groupby(['الادارة التابع لها '])['كود الموظف'].count().reset_index()\
    .sort_values(by='كود الموظف',
     ascending=False)
ax
```

```python
px.bar(data_frame=ax,x='الادارة التابع لها ',y='كود الموظف',color='الادارة التابع لها ',
       text='كود الموظف',title='الكشف بعيادة المطابع')
```

```python
import plotly.express as px

fig = px.bar(data_frame=ax, x='الادارة التابع لها ', y='كود الموظف', 
             color='الادارة التابع لها ',text='كود الموظف')

fig.update_layout(
    plot_bgcolor='rgba(240, 240, 240, 0.8)',
    paper_bgcolor='rgba(255, 255, 255, 0.9)',
    font=dict(
        family="Arial, sans-serif",
        size=14,
        color="black"
    ),
    title=dict(
        text="الرسم البياني للموظفين",
        font=dict(size=20, color='darkblue')
    )
)

fig.show()
```
